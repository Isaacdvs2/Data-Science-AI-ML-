# understand the summary
| Statistic | Meaning                                | Simple Explanation                                             |
| --------- | -------------------------------------- | -------------------------------------------------------------- |
| **count** | Number of non-missing (non-NaN) values | How many data points you have                                  |
| **mean**  | Average value                          | Add them all up and divide by count                            |
| **std**   | Standard deviation                     | Tells how spread out the numbers are (higher = more variation) |
| **min**   | Minimum value                          | The smallest number in the column                              |
| **25%**   | 25th percentile (Q1)                   | 25% of the data is below this value                            |
| **50%**   | Median (Q2)                            | The middle value (half above, half below)                      |
| **75%**   | 75th percentile (Q3)                   | 75% of the data is below this value                            |
| **max**   | Maximum value                          | The largest number in the column                               |


- By default, "df.describe()" only returns the descriptive stats for numerical columns. To include all columns run "df.describe(include='all')

For categorical (text) columns, it shows:

- count: how many non-empty values
- unique: how many different values
- top: most common value
- freq: how often that top value appears
